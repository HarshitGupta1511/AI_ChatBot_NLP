```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load dataset
data = load_iris()
X, y = data.data, data.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Gini Index
gini_clf = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)
gini_clf.fit(X_train, y_train)
gini_pred = gini_clf.predict(X_test)

# Entropy (Information Gain)
entropy_clf = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
entropy_clf.fit(X_train, y_train)
entropy_pred = entropy_clf.predict(X_test)

# Accuracy comparison
gini_acc = accuracy_score(y_test, gini_pred)
entropy_acc = accuracy_score(y_test, entropy_pred)

print(f"Gini Accuracy: {gini_acc:.2f}")
print(f"Entropy Accuracy (ID3): {entropy_acc:.2f}")

# Plot the decision tree
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plot_tree(gini_clf, filled=True, feature_names=data.feature_names)
plt.title("Decision Tree using Gini Index")

plt.subplot(1, 2, 2)
plot_tree(entropy_clf, filled=True, feature_names=data.feature_names)
plt.title("Decision Tree using Entropy (ID3)")
plt.tight_layout()
plt.show()
```
